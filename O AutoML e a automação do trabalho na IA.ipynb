### A necessidade do AutoML
O desempenho de um modelo de Inteligência Artificial depende de vários fatores. Mas o que se deseja é encontrar quais parâmetros da função do modelo conseguem o melhor resultado. No gradiente abaixo, considerando que o modelo possui apenas 3 dimensões – os parâmetros x, y e z – o problema consiste em achar o ponto vermelho no meio do gráfico.


```python
from IPython.display import Image
Image('./Imagens/gradienteponto.jpg')
```

Entretanto, como exemplo, se o modelo utilizar a arquitetura **ResNet18**, o cálculo terá que ser feito com 11.689.512 parâmetros. Encontrar uma configuração de parâmetro ideal é localizar um ponto no espaço de 11.689.512 dimensões. Sem a utilização de **AutoML**, esse ponto pode ser encontrado por tentativas sucessivas nas alterações nos hiperparâmetros (controles externos) e parâmetros (controles internos) dos algoritmos.

O **AutoML** também populariza o aprendizado de máquinas, já que exige um conhecimento muito menor do que o aprendizado de Machine Learning e Deep Learning.

### Como funciona, o exemplo do AutoML-Zero

O *AutoML-Zero,* do Google Research, mostra em gifs o funcionamento genérico dos **AutoMLs**. Utilizando matemática básica, o *AutoML-Zero* automatiza a busca do backpropagation começando com programas vazios ou aleatórios e terminando com a melhor acurácia encontrada.

O objetivo é pesquisar simultaneamente todos os aspectos de um algoritmo de ML, incluindo a estrutura do modelo e a estratégia de aprendizagem, ao mesmo tempo em que emprega o mínimo de interferência humano.

### O WebScraping e o Algoritmo Mestre
Uma profecia diz que somente existirá uma efetiva Inteligência Artificial, quando os programas conseguirem ler o Stack Overflow, site de perguntas e respostas de programação de computadores. O Colab (Google Colaboratory), um ambiente de notebooks Jupyter, que executa gratuitamente nas nuvens códigos em Python direto do navegador, deu o primeiro passo: em caso de erro de programação seu WebScraping indica a consulta ao Stack Overflow.

O AutoML deverá, entretanto, ir mais além. A partir da descrição do problema de negócio a ser resolvido, o WebScraping  deverá procurar na internet os melhores dados para a aprendizagem do modelo.

Outra profecia diz que se o AutoML conseguir incorporar estes 5 algoritmos – árvores de decisão, perceptrons multicamadas, programação genética, o algoritmo Naive Bayes e as SVMs – num único algoritmo será capaz de aprender todas as coisas. Será o Algoritmo Mestre, a última invenção humana.
### A produção
Na outra ponta, o AutoML deverá colocar o modelo em produção em duas etapas separadas e complementares: 1) tornar o modelo disponível para o usuário de maneira visual, isto é, na web e/ou aplicativo; 2) colocar o modelo num computador para que ele possa ser acessado/processado a qualquer momento pela web ou aplicativo acima.

Automatizar a criação de um site básico pode ser feito através de pacotes como o Flask, que é instalado dentro do Python. A criação automática de modelos de aplicativos é uma das possibilidades do GTP-3, do OpenAI-API.

Colocar o modelo dentro de um computador em produção, isso é feito quando se usa o Amazon Web Service, o Google Cloud e o Microsoft Azure, por exemplo.

O sucesso ou não do **AutoML** depende principalmente dos avanços que são feitos nessas duas áreas: aquisição dos dados e facilidade de colocar o modelo em produção. Portanto, o **AutoML**é uma grande parte do futuro do próprio aprendizado de máquina.
